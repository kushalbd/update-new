```python
from bs4 import BeautifulSoup
import requests
page1=requests.get("https://en.wikipedia.org/wiki/Main_Page")#Sends a GET request store in page1 and it contains html content
soup1=BeautifulSoup(page1.content,'html.parser')#parser means raw html into structured object
print(soup1.prettify())#search filter parrts like headings,paragraph, links,etc
```
